```python
class BinaryHeap:
    
    def __init__(self):
        self._heap = []
    
    def _percholate_up(self, current_index):
        
        #WHILE THERE IS A PARENT ITEM.
        while (current_index-1)//2 >= 0:
            parent_index = (current_index - 1)//2 #get parent index
            if self._heap[current_index] < self._heap[parent_index]: #compare values
                #swap if current node's value is smaller than the parent index
                #in order to preserve HEAP PROPERTY (ie, for all node's x, x's parent, p, has a value smaller than x's value)
                self._heap[current_index], self._heap[parent_index] = self._heap[parent_index], self._heap[current_index]
            
            current_index = parent_index
    
    def insert(self, item):
        #append to end
        self._heap.append(item)
        self._percholate_up(len(self._heap) - 1)
        
    def _percholate_down(self, current_index):
        
        # WHILE THERE IS A LEFT CHILD
        # WHY NOT CHECK FOR A RIGHT?
        # BECAUSE IF THERE IS NO LEFT CHID,
        # THEN THERE IS CERTAINLY NO RIGHT CHILD
        while 2 * current_index + 1 < len(self._heap):
            min_child_index = self._get_min_child(current_index)
            if self._heap[current_index] > self._heap[min_child_index]:
                #swap values
                self._heap[current_index], self._heap[minimum_child_index] = self._heap[minimum_child_index], self._heap[current_index]
            
            #else, we are done
            else:
                return
            
            #update current_index
            current_index = min_child_index
    
    def _get_min_child(self, parent_index):
        
        '''This method returns the index that is associated with the
        minimum value for which the parent value should swap with'''
        
        # if the right child's index is GREATER THAN
        # the last index in the list, return left child index
        
        #garret: Still wondering if this is correct
        # if the the index position of where the right child should
        # be is greater than the last element's position
        # we know there is no right child, so pass left child
        if 2*parent_index + 2 > len(self._heap) - 1:
            return 2 * parent_index + 1
        
        # if the left child's value is LESS than the right child's value
        # return the left child's index
        if self._heap[2*parent_index + 1] < self._heap[2*parent_index + 2]:
            return 2 * parent_index + 1
        
        #else, return the right child's index for swapping
        return 2*parent_index + 2
    
    def delete(self):
        #swap the root value and the last element's value
        self._heap[0], self._heap[-1] = self._heap[-1], self._heap[0]
        result = self._heap.pop() #pop/delete off the old root's value
        self._perc_down(0) #move the last element down to it's proper place
        
        return result
                
```
